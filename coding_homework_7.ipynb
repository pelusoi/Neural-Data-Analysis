```python
#Importing relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing data as numpy arrays
url1 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/spikes_left_trial1.csv'
url2 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/spikes_right_trial1.csv'
url3 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/spikes_left_trial2.csv'
url4 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/spikes_right_trial2.csv'

spikes_left_trial1 = np.loadtxt(url1, delimiter = ',')
spikes_right_trial1 = np.loadtxt(url2, delimiter = ',')
spikes_left_trial2 = np.loadtxt(url3, delimiter = ',')
spikes_right_trial2 = np.loadtxt(url4, delimiter = ',')
```

```python
#Printing the length of each array
print(len(spikes_left_trial1))
print(len(spikes_right_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial2))
```

```python
#Determining the firing rate of each trial
print(len(spikes_left_trial1)/30)
print(len(spikes_right_trial1)/30)
print(len(spikes_left_trial2)/30)
print(len(spikes_right_trial2)/30)
```

The results appear to be inconclusive. For trial 1, the firing rate was higher for left turns than right turns. However, for trial 2, the firing rate was higher for right turns than left turns. Perhaps the neuron alternates when it fires.


```python
#Plotting spike train data for each trial in a different plot
plt.subplot(2,1,1)
plt.plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.')
plt.plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.')
plt.xlim([0, 5])
plt.ylim([0, 4])
plt.xlabel('Time [s]')
plt.legend({'Left 1', 'Right 1'})
plt.yticks([])

plt.subplot(2,1,2)
plt.plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.')
plt.plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.')
plt.xlim([0, 5])
plt.ylim([0, 4])
plt.xlabel('Time [s]')
plt.legend({'Left 2', 'Right 2'})
plt.yticks([])
```

The neuron seems more "bursty" during trial 1 than trial 2. The firing data is more spread out over time during the in the first trial than the second trial. This is most likely because the rat learns after trial 1, so the firing data becomes more consistent. When comparing left and right turn trial data, the neuron seems to be more regular-firing during left turn data. The firing data is more consistent and compact for left turn trials in both trial 1 and trial 2. There is a smaller range of time for firing during left trials than during right trials. However, there is more firing overall during right turn trials. There appears to be multiple "bursts" of firing during right turn trials, but only one during left turn trials.

```python
#Calculating inter-spike interval (ISI) arrays for each of the four spike arrays
ISI_left1 = np.diff(spikes_left_trial1)
ISI_right1 = np.diff(spikes_right_trial1)
ISI_left2 = np.diff(spikes_left_trial2)
ISI_right2 = np.diff(spikes_right_trial2)
print(ISI_left1)
print(ISI_right1)
print(ISI_left2)
print(ISI_right2)
```

```python
#Making a histogram for each ISI array
bins=np.arange(0,2,0.01)
plt.hist(ISI_left1, bins=bins)
plt.xlabel('time (s)')
plt.ylabel('Frequency')
plt.title('ISI (left 1)')
plt.show()
bins=np.arange(0,2,0.01)
plt.hist(ISI_right1, bins=bins)
plt.xlabel('time (s)')
plt.ylabel('Frequency')
plt.title('ISI (right 1)')
plt.show()
bins=np.arange(0,2,0.01)
plt.hist(ISI_left2, bins=bins)
plt.xlabel('time (s)')
plt.ylabel('Frequency')
plt.title('ISI (left 2)')
plt.show()
bins=np.arange(0,2,0.01)
plt.hist(ISI_right2, bins=bins)
plt.xlabel('time(s)')
plt.ylabel('Frequency')
plt.title('ISI(right 2)')
plt.show()
```

This neuron does tend to fire in "bursts". Apart from the initial high spikes in activity of the neuron, the left turn data appears to be more spread out and less consistent, with larger gaps between bursts and less activity overall at each burst. The right turn trials tend to have more prominent "bursts" with a large amount of firing in one cluster, then a gap, then another cluster of firing. There appears to be more firing overall in activity during right turn trials.

```python
#Plotting the firing rate over time for each trial
binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_left1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial1[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left1[i] = (len(true_values))/binsize

binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_right1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial1[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right1[i] = (len(true_values))/binsize

plt.plot(fr_bins[:-1], fr_left1, color='red')
plt.plot(fr_bins[:-1], fr_right1, color='blue')
plt.xlabel('Time [s]')
plt.ylabel('Firing Rate [Hz]')
plt.title('Trial 1')
plt.show()

binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_left2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial2[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left2[i] = (len(true_values))/binsize

binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
fr_right2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial2[spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right2[i] = (len(true_values))/binsize

plt.plot(fr_bins[:-1], fr_left2, color='red')
plt.plot(fr_bins[:-1], fr_right2, color='blue')
plt.xlabel('Time [s]')
plt.ylabel('Firing Rate [Hz]')
plt.title('Trial 2')
plt.show()
```


The firing rate between left and right turn trials does differ at specific timepoints during the 30 second waiting period. Overall, it appears that the firing rate for right turn trials occurs at an earlier time in the waiting period than the left turn data for both trials. However, the data is a little bit different between trial 1 and trial 2. For trial 1, the firing rate for both left and right turns appears to spike the highest closer to the middle of the waiting period (15 s), while during trial 2, the right turn trials spike the most at the beginning of the waiting period (3 s) and the left turn trials spike the most at the end of the waiting period (25 s).
