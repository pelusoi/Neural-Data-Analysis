```python
#Importing relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Uploading data files
url1 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter = ',')
url2 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url2, delimiter = ',')
```

```python
#Finding the shape of each array
print(np.shape(acc_hits))
print(np.shape(acc_fa))
```

```python
#Creating the time variable
sampling_rate = 8001/4
time = np.linspace(0, sampling_rate,8001)
```

```python
#Finding the sampling rate of the data
sampling_rate = 8001/4
print(sampling_rate)
```

```python
#plotting the first trial of each data set
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(time, acc_hits[0,])
plt.title('first trial of acc hits')
plt.xlabel('time (s)')
plt.ylabel('LFP during hits (mV)')

plt.subplot(1,2,2)
plt.plot(time, acc_fa[0,])
plt.title('first trial of acc false alarms')
plt.xlabel('time(s)')
plt.ylabel('LFP during false alarms (mV)')
```

For the first trial of acc hits, the dominant oscillation appears to be around 1300 seconds (or Hz). This correlates to an LFP of around -175 mV. For the first trial of acc false alarms, the dominant oscillation appears to be around 900 seconds (or Hz), corresponding to an LFP of around -37 mV. Both of these were gathered based upon the most prominent peak in the spectrum. There is about a difference of 400 Hz between the two data sets, with the first trial of acc hits being longer, while also having a more negative LFP than that of acc false alarms by about 138 mV.


```python
#Plotting ERP of acc_hits
plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage')
plt.title('ERP of acc_hits')
plt.plot([2, 2], [-40,40], 'k', lw=2)
plt.plot([0,2000], [0,0], 'k', lw =2)
```

The graph of ERP data shows more consistent activity in the LFP surrounding screen touch as time goes on. At a lower number of seconds, the voltage is inconsistent, but as the mouse learns when to respond, the voltage peaks are larger, suggesting a more consistent response.

```python
#Plotting ERP of acc_hits and acc_fa
plt.plot(time, acc_fa.mean(0))
plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage')
plt.title('ERP of acc_hits and acc_fa')
plt.plot([2, 2], [-40,40], 'k', lw=2)
plt.plot([0,2000], [0,0], 'k', lw =2)
```

There is a difference in the ERPs between hits and false alarms. Whereas the data for the hits gets more consistent as time goes on with large peaks, the data for false alarms remains variable. This is because the mice are not supposed to respond to the false alarm screen, so the activity in the LFP does not increase during false alarms because the mice are not getting any reward.
