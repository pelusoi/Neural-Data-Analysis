```python
#Importing relevent modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating variables for the data sets
url1 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/lc_hits_hw6.csv'
url2 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/lc_fa_hw6.csv'
url3 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/acc_hits_hw6.csv'
url4 = 'https://raw.githubusercontent.com/pelusoi/Neural-Data-Analysis/main/acc_fa_hw6.csv'

lc_hits =np.loadtxt(url1, delimiter = ',')
lc_fa = np.loadtxt(url2, delimiter = ',')
acc_hits = np.loadtxt(url3, delimiter = ',')
acc_fa = np.loadtxt(url4, delimiter = ',')
```

```python
#Printing the shape of each array
print(np.shape(lc_hits))
print(np.shape(lc_fa))
print(np.shape(acc_hits))
print(np.shape(acc_fa))
```

```python
#Creating the time variable
time = np.linspace(0,4,8001)
```

```python
#Finding the sampling rate
sampling_rate = 8001/4
print(sampling_rate)
```

```python
#plotting the first trials of each data set
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(time, acc_hits[0], color ='red')
plt.plot(time, lc_hits[0], color = 'blue')
plt.xlabel('Time(seconds)')
plt.ylabel('LFP during hits')

plt.subplot(2,1,2)
plt.plot(time, acc_fa[0], color='red')
plt.plot(time, lc_fa[0], color='blue')
plt.xlabel('Time(seconds)')
plt.ylabel('LFP during false alarms')

```


For hits, it appears the dominant oscillation for ACC LFP might be around 9 Hz, while during false alarms the dominant oscillation appears to be around 6 Hz. For the LC LFP, the dominant oscillation appears to be around 10 Hz for the hits and 8 Hz for false alarms, although the peaks are much smaller in the LC than they are in the ACC. Through visual inspection, it appears that the two brain regions have an inverse relationship. When there is a peak in LFP for one region, there is a trough in the other. This relationship appears to be true for both hits and false alarms.

```python
#Computing the coherence with a graph

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

dt = time[1] - time[0]
K = acc_hits.shape[0]
N = lc_hits.shape[1]
ac = np.zeros([2 * N - 1])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real)
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between the acc and lc')
```

In the band of about 7-10 Hz there appears to be the most phase coherence during hits, as that is where there is the largest spike in coherence values. This is the range that the coherence values are highest, indicating that there is the most phase coherence at that frequency.  

```python
#Computing the coherence for hits and false alarms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

dt = time[1] - time[0]
K = acc_hits.shape[0]
N = lc_hits.shape[1]
ac = np.zeros([2 * N - 1])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real, color='red')
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between the acc and lc')

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

dt = time[1] - time[0]
K = acc_fa.shape[0]
N = lc_fa.shape[1]
ac = np.zeros([2 * N - 1])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)
plt.plot(f, cohr.real, color = 'blue')
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between the acc and lc')
```

The frequency at which coherence is strongest between hits and false alarms is very similar. For false alarms the frequency range with the highest coherence is about 7-10 Hz, which is the same as the graph for hits. In general, it appears that coherence for false alarms is a little bit lower than that for hits. If higher coherence means more communication between the ACC and LC, this means that the two brain areas are talking to each other during hits more than during false alarms, which would make sense because a response is needed during a hit but not during a false alarm.
