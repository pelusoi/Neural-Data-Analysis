```python
#Importing relevent modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF
from pandas import DataFrame as df
import pandas as pd
```

```python
#Importing needed files as numpy arrays
fear_traces = np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/fear_conditioning_traces_hw8.csv',delimiter = ',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/avaowitt/NEURO/main/extinction_predictors_hw8.csv',delimiter = ',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int)
```

```python
#Creating the conditioning traces and extinction traces variables with fear traces data.
conditioning_traces = fear_traces[:10458, :]
extinction_traces = fear_traces[10459:, :]
```

```python
#Printing the size and shape of two arrays.
print(np.shape(conditioning_traces))
print(np.shape(extinction_traces))
print(np.size(conditioning_traces))
print(np.size(extinction_traces))
```

```python
#Creating the time_conditioning and time_extinction variables
print(np.shape(fear_traces))
print(24668/15)
time_conditioning = np.linspace(0,697.2,10458)
time_extinction = np.linspace(0,947.26667,14209)
print(10458/15)
print(14209/15)
print(697.2/60)
print(947.26667/60)
print(1644.5333/60)
```

63 ROIs (cells) were imaged during these sessions. The total length of the recording session was 1,644.53 seconds, or 27.41 minutes. The conditioning session had a legnth of 697.2 seconds, or 11.62 minutes. The extinction session had a length of 947.267 seconds, or 15.79 minutes.


```python
#Plotting the first and second ROI values
fig, (ax1, ax2)= plt.subplots(2)

ax1.plot(time_conditioning, conditioning_traces[:,0], color = 'blue')
ax1.plot(time_conditioning, conditioning_traces[:,1], color = 'red' )
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Voltage')
ax1.set_title('Conditioning')
ax1.legend({'ROI1', 'ROI2'})
ax1.plot([shock_index[0], shock_index[0]], [0,1], 'k', lw=1)
ax1.plot([shock_index[1], shock_index[1]], [0,1], 'k', lw=1)
ax1.plot([shock_index[2], shock_index[2]], [0,1], 'k', lw=1)
ax1.plot([tone_index[0], tone_index[0]], [0,1], 'g', lw=1)
ax1.plot([tone_index[1], tone_index[1]], [0,1], 'g', lw=1)
ax1.plot([tone_index[2], tone_index[2]], [0,1], 'g', lw=1)


ax2.plot(time_extinction, extinction_traces[:,0], color = 'blue')
ax2.plot(time_extinction, extinction_traces[:,1], color = 'red')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Voltage')
ax2.set_title('Extinction')
ax2.legend({'ROI1', 'ROI2'})
```

For both conditioning and extinction, the ROI 1 and ROI 2 are related and have similar spikes in voltage. However, ROI 2 seems to have a slightly higher overall voltage for both conditioning and extinction. The conditioning ROIs have wider troughs and spikes than the extinction ROIs, and the extinction ROIs have more "erradic" spikes. The conditioning ROIs have their highest spike in voltage at around 360 seconds, while the extinction ROIs have their highest spike on voltage around 450 seconds.

```python
#Making the predictors_conditioning variable
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
```

```python
#Running a model for each ROI by creating the r_model_conditioning variable using np.zeros

r_model_conditioning = np.zeros(63)
result_list = []

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r3.statistic
  result_list.append(r_model_conditioning)

r_index_conditioning = np.array(result_list)
```

```python
#Creating another loop for the extinction session
shock_array1 = np.zeros(14209)
for i in range(14209):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array1[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array1[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array1[i] = 1
  else:
    shock_array1[i] = 0

tone_array1 = np.zeros(14209)
for i in range(14209):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array1[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array1[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array1[i] = 1
  else:
    tone_array1[i] = 0


predictors_extinction = df(data={'Intercept': np.ones_like(tone_array1), 'tone onset': tone_array1})
predictors_extinction['shock onset'] = shock_array1


r_model_extinction = np.zeros(63)
result_list1 = []

for i in range(63):
  trace_temp1 = extinction_traces[:,i]
  model31 = sm.GLM(trace_temp1, predictors_extinction, family=Poisson())
  model31_results = model31.fit()
  predicted_trace1 = model31_results.predict()
  r31 = pearsonr(trace_temp1, predicted_trace1)
  r_model_extinction[i] = r31.statistic
  result_list1.append(r_model_extinction)

r_index_extinction = np.array(result_list1)
```

```python
#Finding the correlation coefficient
r_index_conditioning1 = r_index_conditioning[~np.isnan(r_index_conditioning)]
r_index_extinction1 = r_index_extinction[~np.isnan(r_index_extinction)]
correlation_coefficient = pearsonr(r_index_conditioning1, r_index_extinction1)
print(correlation_coefficient)
```

```python
#Making a scatter plot
plt.scatter(r_index_conditioning1, r_index_extinction1)
plt.xlabel('r_index_conditioning')
plt.ylabel('r_index_extinction')
```

The correlation coefficient is about 0.44, which is a somewhat strong correlation (perfect correlation would be 1). The scatterplot also shows this, as the data has a somewhat linear relationship, although it is a quite spread out line. This means that a cell that responds stongly to shock and tone onset during conditioning is moderately more likely to respond strong to tone onset during the extinction session.
